# 1. Basic DataFrame Operations

```python
import numpy as np
import pandas as pd
import random
```

```python
names = [
    "Dhavall", "Juicy", "Abi", "Herle", "Varda", "Vishy", "Aagy", "Rasgoti", "Rane", "Vegand",
    "Ishu", "Sampya", "Takya", "Mohini", "San", "Arya"
]

subjects = [
    "AppMaths", "FluidFlow", "CEOps", "Thermo", "CRE", "HT", "AppChem", "SepPro", "ProSft", "EVS"
]

marks = {name: [np.random.randint(100) for _ in range(len(subjects))] for name in names}

for name, scores in marks.items():
    print(f"{name}: {dict(zip(subjects, scores))}")
```

# 4. Grouping and Aggeration

```python
names = ['dhavall', 'juicy', 'abi', 'herle', 'varda', 'vishy', 'aagy', 'rasgoti', 'rane', 'vegand', 'ishu', 'sampya', 'takya', 'mohini', 'san', 'arya']
subjects = ['AppMaths', 'FluidFlow', 'CEOps', 'Thermo', 'CRE', 'HT', 'AppChem', 'SepPro', 'ProSft', 'EVS']
branches = ['Chemical', 'Mathematics', 'Science', 'Mathematics', 'Chemical', 'Science', 'Time Waste', 'Chemical', 'Time Waste', 'Science',
            'Chemical', 'Mathematics', 'Science', 'Mathematics', 'Chemical', 'Science']
cgpas = [9.4, 9.3, 9.2, 9.1, 9.0, 8.9, 8.8, 8.7, 8.6, 8.5, 9.35, 8.95, 8.75, 9.05, 8.85, 9.15]
random.shuffle(cgpas)

marks = {name: [np.random.randint(100) for _ in range(10)] for name in names}

data = {'Name': names, 'Branch': branches, 'CGPA': cgpas}
df = pd.DataFrame(data)

df_marks = pd.DataFrame(marks, index=subjects).T

display(df)
# display(df_marks)
```

```python
df.groupby(df['CGPA']).mean()
```

```python
df.pivot_table( values= 'CGPA',index='Branch' ,aggfunc = 'mean')
```

```python
df.groupby('Branch').sample()
```

# 5. Merging and Joining Data

```python
data1 = {'Roll': [151,152,153,154,156,157,158,159,160],
        'Names': ['Ayush','Parin','Siddhi','Sourabh','Dhavall','Aarya','Yashraj','Karan','Soham']}

data2 = {'Roll': [151,152,153,154,156,157,158,159,160],
        'CGPA':  [5.82,8.36,9.36,7.73,8.64,7.36,7.27,6.18,7.18]}

df_name = pd.DataFrame(data1)
df_cgpa = pd.DataFrame(data2)
```

```python
pd.concat([df_name, df_cgpa], ignore_index = True)
```

```python
pd.merge(df_name, df_cgpa) # Looks at the common columns and merges them
```

```python
pd.merge(df_name, df_cgpa, how = 'inner') # Well IDK What is the Difference here
```
