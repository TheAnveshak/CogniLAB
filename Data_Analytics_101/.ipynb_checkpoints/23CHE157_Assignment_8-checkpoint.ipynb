```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

ds = sns.load_dataset('penguins')
```

```python
ds
```

```python
ds.shape
ds.info()
ds.describe()
ds.isnull().sum()
```

```python
ds['island'].unique()
ds['sex'].unique
ds['species'].unique()
```

```python
species = ds['species'].value_counts()
species.plot(kind='pie',autopct="%.3f%%")
iland = ds['island'].value_counts()
iland.plot(kind='pie',autopct="%.3f%%")
sex = ds['sex'].value_counts()
sex.plot(kind='pie',autopct="%.3f%%")
```

```python
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", data=ds, hue="species")
plt.title("Bill Length v/s Bill Depth", size=20, color="Magenta")
```

```python
sns.histplot(x ="body_mass_g", data=ds, kde = True, hue = "species")
plt.title("Mass", size=20, color="red")
```

```python
sns.boxplot(x = "species", y = "flipper_length_mm", data =ds, hue = "sex")
plt.title("Flipper Length of 3 species wrt Sex", size=20, color="red")
# why is there o
```

```python
sns.pairplot(ds, hue = "species", height=3, diag_kind="hist")
```
