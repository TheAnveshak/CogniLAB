```python
alive = ds['alive'].value_counts()
alone = ds['alone'].value_counts()
town = ds['embark_town'].value_counts()
sex = ds['sex'].value_counts()
age = ds['age'].value_counts()
clas = ds['class'].value_counts()
fare = ds['fare'].value_counts()
```

```python
sns.scatterplot(x='fare', y='age', data=ds, hue="alive")
plt.title("Age v/s Fare", size=20, color="Magenta")
plt.show()
```

```python
sns.countplot(x="pclass", hue="survived", data=ds)
plt.title("Class vs Survival", size=20, color="Magenta")
plt.show()
```

```python
sns.pairplot(ds[['age', 'fare', 'survived', 'pclass']])
plt.show()
```
